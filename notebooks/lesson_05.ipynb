You are a nutritionist researching two different types of diets to see if there is a significant difference in weight loss after one month. You choose two random groups of people; one group follows the first diet, and the other follows the second. At the end of the month, you record the weight loss (in kg) for each person:

With these data, it seeks to answer the following question: Is there a significant difference in average weight loss between people who followed the first diet and those who followed the second diet?

To draw conclusions, follow the points below:

State the hypothesis: null and alternative hypothesis.
Perform the test to test the hypothesis. You can use a Student's t-test.
Analyze the conclusions.

```python
#menos de 30 muestras no se puede usar una prueba "z" (z_score) y se usa la prueba "t" (destribucion de tdestude)
# Hipotesis nula_0: mu_1 = mu_2 
# Hipotesis alterna h_1 = mu_1 <> mu_2

diet_1 = [2.0, 2.5, 3.0, 2.8, 2.3, 2.7, 2.5]
diet_2 = [3.0, 3.2, 3.1, 2.9, 2.8, 3.0, 3.2]

mean_diet_1 = np.mean(diet_1)
mean_diet_2 = np.mean(diet_2)

print(mean_diet_1, mean_diet_2)

t_value, p_value = stats.ttest_ind(diet_1,diet_2)
print(p_value)

# print(mean_diet_1)
```

## Problem 2
A farmer decides to test three different types of fertilizers to determine if one is superior in terms of corn production. The farmer plants corn on 15 identical plots and uses all three fertilizers (5 plots for each type). At the end of the season, he measures the corn yield (in kg) of each plot, with the following result:

With these data, it seeks to answer the following question: Is there a significant difference in average weight loss between people who followed the first diet and those who followed the second diet?

To draw conclusions, follow the points below:

State the hypothesis: null and alternative hypothesis.
Perform the test to test the hypothesis. You can use a Student's t-test.
Analyze the conclusions.

```python
fertilizer_1 = [20, 21, 20, 19, 20]
fertilizer_2 = [22, 21, 23, 22, 21]
fertilizer_3 = [24, 23, 22, 23, 24]

print(np.mean(fertilizer_1), np.mean(fertilizer_2), np.mean(fertilizer_3))

# Hipotesis nula no nay diferencia (<>) en los promedios de rendimiento de cada fertilizante
# Hipotesis alterna siempre va a ser asi cuando sean 3 o mas, hay diferencias en al menos 
# dos tipos de fertilizantes y nuestro nivel de significancia alfa es 0.05
# Se calcula el f_staticts

f_statistic, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(p_value)
```

Una compañia industrial a adquirido tres maquinas nuevas de diferentes marcas y desea determinar si una de ellas es mas rapida que las otras. Cifras de cinco horas de produccion de cada maquina son observadas al azar.
Que conclusiones pueden extraerse en alfa = 00.5? Y en alfa = 0.01?

```python
mac_1 = [25,30,36,38,31]
mac_2 = [31,39,38,42,35]
mac_3 = [24,30,28,25,28]

# Hipotesis nula no hay diferencia entre la produccion de las maquinas.
# Hipotesis alterna hay diferencia por lo menos entre 2 de las maquinas en su produccion.

print(np.mean(mac_1), np.mean(mac_2), np.mean(mac_3))

f_statistic, p_value = stats.f_oneway(mac_1, mac_2, mac_3)

print(p_value)

# Alfa = 0.05
# Alfa = 0.01
# Improbable, al menos dos maquinas son diferentes.

t_value, p_value = stats.ttest_ind(mac_1, mac_2)
print(p_value)
# Si alfa es mayor que el p_value entonces aceptamos h_1, si no rechazamos h_1.
# Hipotesis nula las 2 maquinas son iguales.
# Hipotesis alterna las 2 maquinas son diferentes.
```

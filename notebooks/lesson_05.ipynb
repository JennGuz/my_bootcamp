```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("/workspaces/my_bootcamp/data/ACCIDENTS_GU_BCN_2010.csv", encoding="latin-1")
df
```

```python
df.info()
```

```python
df["Date"] = df[u"Dia de mes"].apply(lambda x: str(x)) + "-" + df[u"Mes de any"].apply(lambda x: str(x)) + "-2010"
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")
df.sample(6)
```

```python
accidents = df.groupby(["Date"]).size()
print("Accidents", accidents.mean())
```
